# 1. WAGES 데이터 분석

```r
wages <- read.csv("DecisionTree/wages-1.csv", header = T)
```

```r
str(wages)
```

1084 x 20의 데이터가 존재한다. <br>
임금에 자연 로그가 씌워진 연속형 변수 lnwage를 target으로 회귀분석과 의사결정나무 회귀분석을 실시한다.

```r
wages[,2]=as.factor(wages[,2])
wages[,3]=as.factor(wages[,3])
wages[,4]=as.factor(wages[,4])
wages[,5]=as.factor(wages[,5])
wages[,6]=as.factor(wages[,6])
wages[,7]=as.factor(wages[,7])
wages[,10]=as.factor(wages[,10])
wages[,13]=as.factor(wages[,13])
wages[,14]=as.factor(wages[,14])
wages[,15]=as.factor(wages[,15])
wages[,16]=as.factor(wages[,16])
wages[,17]=as.factor(wages[,17])
wages[,18]=as.factor(wages[,18])
wages[,19]=as.factor(wages[,19])
wages[,20]=as.factor(wages[,20])
```

0과 1의 이진값을 가진 변수들을 팩터화 시킨다.

```r
set.seed(42)
ind <- sample(2, nrow(wages), replace = TRUE, prob = c(0.7, 0.3))
wages.train <- wages[ind == 1, ]
wages.test <- wages[ind == 2, ]
```

train : test = 7 : 3 비율로 데이터를 분리하여 준다.

```r
wages.rpart <- rpart(lnwage~., data = wages.train,
                     control = rpart.control(minsplit = 10))
```

최소 노드 수를 10으로 하여 의사결정나무 회귀 모델을 생성한다.

```r
wages.rpart$cptable
```

각 노드 별 CP와 오분류율은 위와 같다.

```r
wages.rpart
```

```r
rpart.plot(wages.rpart)
```

cps85는 조사 시점으로 85년 이후 조사인지 이전 조사인지 나타내는 변수이다. 85년 시점이 no인 이전 조사의 가지에서는 gender가 더 높은 영향을 미치고 85년 조사에서는 성별보다는 교육 수준이 더 큰 영향을 끼친다는 해석이 가능하다.

```r
wages.rpart.pred <- predict(wages.rpart, newdata = wages.test)
wages.xlim <- range(wages$lnwage)
plot(wages.rpart.pred ~ lnwage, data = wages.test, xlab = "Observed",
    ylab = "Predicted", ylim = wages.xlim, xlim = wages.xlim)
abline(a = 0, b = 1)
```

점들이 마치 범주처럼 위치해있는데 이는 변수를 범주화시키는 의사결정나무의 특징이다.

```r
wages.reg <- lm(lnwage~., data = wages.train)
```

```r
summary(wages.reg)
```

결정계수는 47.32%이다. <br>
f검정 결과 p-value가 2.2e-16으로 매우 낮아 적어도 하나의 유의미한 계수가 있다. <br>
t검정 결과 유의미한 계수는 절편, hispanic1, gender1, expersq, union1, manuf1, constr1, manag1, service1, prof1, cps851이다.

```r
par(mfrow=c(2,2))
plot(wages.reg)
```

Residual plot과 Q-Q plot을 보았을 때 분산의 동질성과 선형성은 만족하는 것 같으나 아웃라이어가 존재하는 것으로 보여짐

```r
wages.reg.pred <- predict(wages.reg, newdata = wages.test)
```

```r
plot(wages.test$lnwage, wages.reg.pred)
abline(a=0, b=1)
```

#  hmeq 데이터

```r
hmeq.data <- read.csv("Regression/hmeq.csv", header = T)
```

```r
str(hmeq.data)
```

5960 x 13의 데이터가 있다. <br>
BAD를 target 변수로 하여 분석을 실시한다.

```r
set.seed(1234) 
ind <- sample(2, nrow(hmeq.data), replace=TRUE, prob=c(0.7, 0.3))
hmeq.train <- na.omit(hmeq.data[ind == 1,])
hmeq.test <- na.omit(hmeq.data[ind==2,])
```

train과 test를 7:3으로 분리한다.

## (1)  로지스틱회귀모형 적합과 오즈비의 해석

```r
hmeq.log <- glm(BAD~., family = binomial, data = hmeq.train)
```

```r
summary(hmeq.log)
```

```r
hmeq.step <- step(hmeq.log, direction = "both")
```

```r
summary(hmeq.step)
```

stepwise 방법으로 변수를 선택하였다.

```r
exp(hmeq.step$coefficients)
```

오즈비를 보았을 때 JOBSales가 한 단위 증가하면 BAD로 예측될 확률이 30배 가량 증가한다.

## (2) 예측 사후확률 p 측정 

```r
hmeq.train.p <- predict(hmeq.step, hmeq.train, type="response")
hmeq.test.p <- predict(hmeq.step, newdata = hmeq.test, type="response")
```

```r
head(hmeq.train.p)
```

train 데이터에 대한 사후 확률이 위와 같이 나왔다.

```r
table(hmeq.train$BAD, hmeq.train.p > 0.5)
```

```r
(2271 + 59) / (152 + 12 + 2271 + 59) #정확도
59 / (152 + 59) #특이도
2271 / (2271 + 12) #민감도
```

임계값을 0.5로 잡았을 때 분류 결과는 위와 같다. <br>
신용이 나쁜 사람을 나쁘다고 판단하는 특이도가 매우 낮다.

```r
table(hmeq.train$BAD, hmeq.train.p > 0.1)
```

```r
(1898 + 136) / (1898 + 136 + 385 + 75) #정확도
136 / (136 + 75) #특이도
1898 / (1898 + 385) #민감도
```

임계값을 0.1로 잡았을 때 분류 결과는 위와 같다. <br>
0.5보다 정확도와 민감도는 떨어졌지만 특이도가 많이 상승하였다. 따라서 test 모델에는 임계값을 0.1로 적용한다.

```r
table(hmeq.test$BAD, hmeq.test.p > 0.1)
```

```r
(761 + 58) / (761 + 58 + 162 + 40) #정확도
58 / (58 + 40) #특이도
761 / (761 + 162) #민감도
```

테스트에 적용한 결과는 위와 같다. 모델이 잘 적합되었는지 지표가 train과 크게 차이나지 않는다.

# Housing 데이터 

```r
housing.data <- read.csv("Regression//housing.csv", header = T)
```

```r
str(housing.data)
```

```r
housing.data$CHAS=as.factor(housing.data$CHAS)
```

506 X 14 데이터가 존재한다. <br>
MEDV를 target 변수로 한다.

## (1) 의사결정나무

```r
housing.rpart <- rpart(MEDV ~., data = housing.data, control = rpart.control(minsplit = 10))
```

```r
rpart.plot(housing.rpart)
```

RM, LSTAT, CRIM, DIS, PTRATIO가 분기 역할을 했다.

```r
housing.rpart.pred <- predict(housing.rpart)
plot(housing.rpart.pred ~ MEDV, data = housing.data)
abline(a=0, b=1)
```

## (2) 선형회귀

의사결정나무에서 RM, LSTAT, CRIM, DIS, PTRATIO가 분기 역할을 했다. 이를 바탕으로 우선 선형식을 세워본다.

```r
linearFormula <- MEDV ~ RM + LSTAT + CRIM + DIS + PTRATIO
```

```r
housing.linear <- lm(linearFormula, data = housing.data)
```

```r
summary(housing.linear)
```

```r
par(mfrow = c(2, 2))
plot(housing.linear)
```

선형 회귀에 적합하지 않다. residual 플롯이 경향성을 띄고 Q-Q플롯을 보았을 때 선형성을 만족하지 못 하는 것으로 보인다.

```r
unlinearFormula <- MEDV ~ RM + I(RM ^ 2) + LSTAT + I(LSTAT * RM) + CRIM + DIS + PTRATIO
```

다항식을 추가하여 다항회귀를 실시한다.

```r
housing.unlinear <- lm(unlinearFormula, data = housing.data)
```

```r
summary(housing.unlinear)
```

```r
par(mfrow = c(2, 2))
plot(housing.unlinear)
```

Resiual 플롯을 보았을 때 선형회귀를 했을 때보다 경향성이 사라지고 0을 중심으로 퍼진다. Q-Q플롯을 보았을 때도 선형 회귀보단 직선에 적합한다. 그러나 아웃라이어가 심해졌고 아직 선형성을 만족한다고 볼 수는 없을 것으로 보인다.
